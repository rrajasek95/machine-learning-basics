# Linear Regression and Basis Functions

Linear Regression is a modelling technique that tries to map input $\bf{x}$ to a continuous target variable $t$ using a hypothesis $y({\bf x} , {\bf w })$.

A simple linear regression hypothesis is of the form:

$$ y({\bf x},{\bf w}) = w_0 + w_1 x_1 + \dots + w_Dx_D $$

Where ${\bf x} = (x_1,\dots,x_D)^T$.

However, this form is not very useful as it's linear on the components, restricting us to only hyperplane models.

We can generalize linear regression to non-linear functions by defining a class of functions $\phi$ on ${\bf x}$ that map to a scalar. We express our hypothesis as follows:

$$
\begin{align}
    y({\bf x},{\bf w}) & = w_0 + w_1 \phi_1({\bf x}) + \dots + w_{M - 1} \phi_{M - 1}({\bf x}) \\
    y({\bf x},{\bf w}) & = w_0 + \sum_{j = 1}^{M - 1}\phi_{j}({\bf x})
\end{align}
$$

These $\phi_j({\bf x})$ are called basis functions. We have $M$ parameters $(w_0,\dots,w_{M - 1})$ in the model.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4, 0.1).T
# Assuming that our model has some implicit noise
t = np.sin(np.pi * x) # + np.random.normal(scale=.1, size=x.shape)

plt.scatter(x, t)
plt.show()
```

```python
mu = np.arange(0.0, 4.0, 0.2)
s = 0.1


sinv = -1/(2*s**2)
phi = np.array([np.exp((i - mu)**2/sinv) for i in x])

w_ML = np.matmul(np.linalg.pinv(phi), t.T)

plt.scatter(x, t, label='Input Data')
plt.plot(x, np.matmul(phi, w_ML), label='Hypothesis')
plt.legend()
plt.show()

print(w_ML)
```
