# Linear Regression and Basis Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.1).T
y = np.sin(2*np.pi * x)

print(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def x_basis(x):
    return x

def generate_weights(n):
    return np.random.normal(size=n).reshape((n,1))

def compute_gaussian(x, mu, s):
    return np.exp((x - mu)**2 / (2. * s ** 2))
```

```python
num_parameters = 3

w = generate_weights(num_parameters)

mu_vec = np.random.normal(size=num_parameters - 1)
s_vec = np.random.normal(loc=2, size=num_parameters - 1)

def compute_y(x, w, mu_vec, s_vec):
    g = compute_gaussian(x, mu_vec, s_vec).reshape(mu_vec.shape[0], 1)
    
    return np.asscalar(w[0] + np.dot(w[1:].T, g))

y_h = list(map(lambda x_: compute_y(x_,w,  mu_vec, s_vec), x))

plt.plot(x, y_h)
plt.plot(x, y)
plt.show()
```
